```python
import pandas as pd
env_policies = pd.read_csv('..\Powering_Recycling_in_Spain\Datasets\ENVIROMENTAL_POLICY.csv')
env_policies = env_policies.round(2)
env_policies
```

```python
list(env_policies['Value'].unique())
```

```python
env_policies['Country'].unique()
```

```python
selected_policy = env_policies.loc[env_policies['Country'].isin(['Spain', 'Australia'])].copy()
selected_policy = selected_policy[selected_policy['Year'] > 2010].copy()
selected_policy = selected_policy[selected_policy['Variable'].isin(['Environmental Policy Stringency'])].copy()
columns_to_drop = ['COU', 'VAR', 'YEA', 'Unit Code', 'Unit', 'PowerCode Code', 'PowerCode', 'Reference Period Code', 'Reference Period', 'Flag Codes', 'Flags']
selected_policy.drop(columns=columns_to_drop, inplace=True)
selected_policy
```

```python
def new_df_country(country):
    selected_policy_df = selected_policy[selected_policy['Country'] == country].copy()
    selected_policy_df_ = selected_policy_df.transpose()
    selected_policy_df_.reset_index(drop=True, inplace=True)
    rows_to_delete = [0, 1]
    selected_policy_df_.drop(rows_to_delete, inplace=True)
    selected_policy_df_.columns = selected_policy_df_.iloc[0]
    selected_policy_df_= selected_policy_df_.iloc[1:]
    return selected_policy_df_
```

```python
df_Spain = new_df_country('Spain')
```

```python
df_Australia = new_df_country('Australia')
```

```python
result_df = pd.concat([df_Australia, df_Spain], ignore_index=True)
result_df
```

```python
result_df['Country'] = ['Australia', 'Spain']
result_df
```

```python
result_df.reset_index(drop=True, inplace=True)
result_df.set_index('Country', inplace=True)
```

```python
result_df
```

```python
result_df.columns.name = None
```

```python
result_df.reset_index(inplace=True)
```

```python
result_df
```

```python
numeric_cols = result_df.columns[1:]
new_row = result_df[numeric_cols].iloc[0] - result_df[numeric_cols].iloc[1]
result_df.loc[result_df.index[-1] + 1] = result_df.iloc[-1].copy()
result_df.loc[result_df.index[-1], numeric_cols] = new_row.values
result_df
```

```python
numeric_cols = result_df.columns[1:]
new_row = (result_df[numeric_cols].iloc[2] / result_df[numeric_cols].iloc[1])*100
result_df.loc[result_df.index[-1] + 1] = result_df.iloc[-1].copy()
result_df.loc[result_df.index[-1], numeric_cols] = new_row.values
result_df
```

```python
result_df.loc[2, 'Country'] = 'Difference'
result_df.loc[3, 'Country'] = 'Difference in %'
result_df
```

```python
result_df.iloc[3, 1:] = pd.to_numeric(result_df.iloc[3, 1:], errors='coerce').round(2)
```
