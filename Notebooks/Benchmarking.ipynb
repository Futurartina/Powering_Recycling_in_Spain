```python
import pandas as pd
```

```python
happiness = pd.read_csv(r'..\CSV\benchmarking_happiness.csv')
policies = pd.read_csv(r'..\CSV\benchmarking_policies.csv')
EPI = pd.read_csv(r'..\CSV\benchmarking_EPI.csv')
```

```python
merged_benchmarking = pd.merge(happiness, policies, on='Country', how='left')
merged_benchmarking = pd.merge(merged_benchmarking, EPI, on='Country', how='left')
merged_benchmarking
```

```python
merged_benchmarking_filtered = merged_benchmarking[['Country', 'Recycling', '2020', 'Logged GDP per capita', 'Perceptions of corruption', 'Lead exposure', 'Unsafe drinking water']]
merged_benchmarking_filtered
```

```python
merged_benchmarking_filtered = merged_benchmarking_filtered.rename(columns={'2020': 'Environmental policies', 'Logged GDP per capita': 'GDP per capita', 'Perceptions of corruption': 'Perception of corruption'})
merged_benchmarking_filtered
```

```python
merged_benchmarking_filtered.to_csv(r"..\CSV\merged_benchmarking.csv", index=False)
```
