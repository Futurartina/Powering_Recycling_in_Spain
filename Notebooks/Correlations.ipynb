```python
import pandas as pd
```

```python
EPI = pd.read_csv(r"..\CSV\EPI.csv")
df_4square = pd.read_csv(r"..\CSV\4square.csv")
env_policies_gral = pd.read_csv(r"..\CSV\env_policies_gral.csv")
Hapiness_Index = pd.read_csv(r"..\CSV\Hapiness_Index.csv")
Literacy__ = pd.read_csv(r"..\CSV\Literacy__.csv")
median_age = pd.read_csv(r"..\CSV\median_age.csv")
pop_dens = pd.read_csv(r"..\CSV\pop_dens.csv")
```

```python
merged_df = pd.merge(EPI, df_4square, on='Country', how='left')
merged_df = pd.merge(merged_df, env_policies_gral, on='Country', how='left')
merged_df = pd.merge(merged_df, Hapiness_Index, on='Country', how='left')
merged_df = pd.merge(merged_df, Literacy__, left_on='Country', right_on='Country Name', how='left')
merged_df = pd.merge(merged_df, median_age, on='Country', how='left')
merged_df = pd.merge(merged_df, pop_dens, on='Country', how='left')
```

```python
columns_to_drop = ['Country', 'Latitude', 'Longitude', 'Unnamed: 0', 'Country Name', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
merged_df.drop(columns=columns_to_drop, inplace=True)
new_column_name = {'Value': 'Environmental_policies'}
merged_df = merged_df.rename(columns=new_column_name)
merged_df

```

```python
correlations = merged_df.corr()['Recycling']
sorted_correlation = correlations.sort_values(ascending=False)
sorted_correlation
```
