```python
sorted_EPI = EPI.sort_values(by='Recycling.new', ascending=False)
sorted_EPI.head(10)
```

```python
# In order to define which country to benchmark, it is necessary to rank them according to the "Recycling" value. 
# South Korea is number 1 but it is very difficult to find information about it. 
# I have decided on Australia because it is one of the best in recycling management and it is easier to find information. 
```

```python
# Then, which are the values of the defined variables for Australia and Spain?
# What are the differences?
```

```python
selected_columns = ['country', 'Recycling.new', 'Waste Management.new', 'Environmental Health.new', 'Sanitation & Drinking Water.new', 'Lead exposure.new', 'Heavy Metals.new', 'Unsafe drinking water.new', 'Air Quality.new', 'Unsafe sanitation .new', 'Water Resources.new', 'Wastewater treatment.new', 'PM2.5 exposure.new', 'Controlled solid waste.new', 'Household solid fuels.new', 'Ocean Plastics.new']
EPI_correlated = sorted_EPI[selected_columns]
```

```python
EPI_correlated
```

```python
EPI_correlated = EPI_correlated.copy()
columns_to_drop = ['Waste Management.new', 'Environmental Health.new', 'Sanitation & Drinking Water.new', 'Heavy Metals.new', 'Air Quality.new', 'Water Resources.new']
EPI_correlated.drop(columns=columns_to_drop, inplace=True)
EPI_correlated.rename(columns={'country': 'Country'}, inplace=True)
EPI_correlated.columns = EPI_correlated.columns.str.replace('.new', '')
EPI_correlated
```

```python
EPI_correlated.to_csv(r"..\CSV\EPI.csv", index=False)
```

```python
selected_rows = EPI_correlated.loc[EPI_correlated['Country'].isin(['Spain', 'Australia'])].copy()
selected_rows
```

```python
numeric_cols = selected_rows.columns[2:]
new_row = selected_rows[numeric_cols].iloc[0] - selected_rows[numeric_cols].iloc[1]
selected_rows.loc[selected_rows.index[-1] + 1] = selected_rows.iloc[-1].copy()
selected_rows.loc[selected_rows.index[-1], numeric_cols] = new_row.values
selected_rows
```

```python
selected_rows.loc[53, 'Country'] = 'Difference'
selected_rows
```

```python
numeric_cols = selected_rows.columns[2:]
new_row = (selected_rows[numeric_cols].iloc[2] / selected_rows[numeric_cols].iloc[1]).round(2)*100
selected_rows.loc[selected_rows.index[-1] + 1] = selected_rows.iloc[-1].copy()
selected_rows.loc[selected_rows.index[-1], numeric_cols] = new_row.values
selected_rows
```

```python
selected_rows.loc[54, 'Country'] = 'Difference in %'
selected_rows
```

```python
selected_rows.to_csv(r"..\CSV\benchmarking_EPI.csv", index=False)
```
