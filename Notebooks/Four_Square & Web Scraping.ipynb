```python
load_dotenv()
token = os.getenv("token")
```

```python
def requests_for_foursquare (query, lat, lon, radius):
    url = f"https://api.foursquare.com/v3/places/search?query={query}&ll={lat}%2C{lon}&radius={radius}"
    
    headers = {
        "accept": "application/json",
        "Authorization": token
        }  
    response = requests.get(url, headers=headers)
    return response
```

```python
element = requests_for_foursquare ("Recycling Facility", -35.2834600, 149.1280700, 10000)
```

```python
res = element.json()
res
```

```python
import numpy as np
def extracting_from_one_element (element):
    name = element["name"]
    distance = element["distance"]
    lat = element["geocodes"]["main"]["latitude"]
    lon = element["geocodes"]["main"]["longitude"]
    
    if "timezone" in list(element.keys()):
        region = element["timezone"]
    else:
        region = np.nan
        
    
    small_dict = {
    "name": name,
    "distance": distance,
    "lat": lat,
    "lon": lon,
    "region": region
}
    return small_dict
```

```python
recycling_facilities = []
for i in res["results"]:
    recycling_facilities.append(extracting_from_one_element(i))
recycling_facilities
```

```python
recycling_facilities = pd.DataFrame(recycling_facilities)
recycling_facilities["Type"] = "Recycling Facility"
recycling_facilities
```
