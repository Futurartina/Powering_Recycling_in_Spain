```python
import pandas as pd
```

```python
happiness = pd.read_excel('..\Datasets\WORLD_HAPPINESS.xls')
happiness
```

```python
columns_to_drop = list(range(2, 5)) + list(range(happiness.shape[1] - 7, happiness.shape[1]))
happiness = happiness.drop(columns=happiness.columns[columns_to_drop])
happiness
```

```python
happiness = happiness.rename(columns={'Country name': 'Country'})
happiness
```

```python
happiness.to_csv(r"..\CSV\Hapiness_Index.csv", index=False)
```

```python
selected_happiness = happiness.loc[happiness['Country'].isin(['Spain', 'Australia'])].copy()
numeric_cols = selected_happiness.columns[1:]
new_row = selected_happiness[numeric_cols].iloc[0] - selected_happiness[numeric_cols].iloc[1]
selected_happiness.loc[selected_happiness.index[-1] + 1] = selected_happiness.iloc[-1].copy()
selected_happiness.loc[selected_happiness.index[-1], numeric_cols] = new_row.values
selected_happiness.loc[32, 'Country'] = 'Difference'
numeric_cols_ = selected_happiness.columns[1:]
new_row_ = (selected_happiness[numeric_cols_].iloc[2] / selected_happiness[numeric_cols_].iloc[1]).round(2)*100
selected_happiness.loc[selected_happiness.index[-1] + 1] = selected_happiness.iloc[-1].copy()
selected_happiness.loc[selected_happiness.index[-1], numeric_cols_] = new_row_.values
selected_happiness.loc[33, 'Country'] = 'Difference in %'
selected_happiness
```

```python
selected_happiness.to_csv(r"..\CSV\benchmarking_happiness.csv", index=False)
```
