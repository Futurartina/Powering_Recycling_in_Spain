```python
import pandas as pd
```

```python
literacy = pd.read_excel("..\Datasets\LITERACY.xls")
literacy
```

```python
literacy = literacy.drop([0, 1])
literacy
```

```python
literacy.columns = literacy.iloc[0]
literacy = literacy.drop([2])
literacy
```

```python
literacy = literacy.iloc[:, [0] + list(range(len(literacy.columns)-10, len(literacy.columns)))]
literacy
```

```python
literacy.columns = literacy.columns.to_series().apply(lambda x: int(float(x)) if pd.notnull(x) and x != literacy.columns[0] else x)
literacy.columns.name = None
literacy = literacy.reset_index(drop=True)
literacy
```

```python
literacy.to_csv((r"..\CSV\Literacy.csv"))
```
