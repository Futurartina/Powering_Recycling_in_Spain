```python
import pandas as pd
```

```python
median = pd.read_csv("..\Datasets\MEDIAN_AGE.csv")
median
```

```python
median_columns = ['name', 'value']
median = median[median_columns].copy()
median.rename(columns={'name': 'Country', 'value': 'Median Age'}, inplace=True)
median
```

```python
median.to_csv(r"..\CSV\median_age.csv", index=False)
```

```python
median_columns = ['name', 'value', 'date_of_information']
median = median[median_columns].copy()
median = median.loc[median['name'].isin(['Spain', 'Australia'])].copy()
median
```

```python
temp_row = median.iloc[1].copy()  
median.iloc[1] = median.iloc[0]  
median.iloc[0] = temp_row
median
```

```python
new_columns_names = ['Country', 'Median_Age', 'Date_of_info']
median.columns = new_columns_names
```

```python
median
```

```python
numeric_cols = median.columns[1:]
new_row = median[numeric_cols].iloc[0] - median[numeric_cols].iloc[1]
median.loc[median.index[-1] + 1] = median.iloc[-1].copy()
median.loc[median.index[-1], numeric_cols] = new_row.values
median
```

```python
numeric_cols = median.columns[1:]
new_row = (median[numeric_cols].iloc[2] / median[numeric_cols].iloc[1])*100
median.loc[median.index[-1] + 1] = median.iloc[-1].copy()
median.loc[median.index[-1], numeric_cols] = new_row.values
median
```

```python
median.loc[70, 'Country'] = 'Difference'
median.loc[71, 'Country'] = 'Difference in %'
median
```
