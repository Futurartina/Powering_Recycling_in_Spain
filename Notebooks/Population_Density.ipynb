```python
import pandas as pd
```

```python
pop = pd.read_csv("../Datasets/POPULATION_DENSITY.csv")
pop
```

```python
pop_dens_columns = ['country', 'pop2023', 'area', 'density']
pop_dens = pop[pop_dens_columns].copy()
pop_dens = pop_dens.loc[pop_dens['country'].isin(['Spain', 'Australia'])].copy()
pop_dens
```

```python
temp_row = pop_dens.iloc[1].copy()  
pop_dens.iloc[1] = pop_dens.iloc[0]  
pop_dens.iloc[0] = temp_row 
```

```python
pop_dens
```

```python
new_columns_names = ['Country', 'Population', 'Area(km2)', 'Density(pop/km2)']
pop_dens.columns = new_columns_names
```

```python
pop_dens
```

```python
numeric_cols = pop_dens.columns[1:]
new_row = pop_dens[numeric_cols].iloc[0] - pop_dens[numeric_cols].iloc[1]
pop_dens.loc[pop_dens.index[-1] + 1] = pop_dens.iloc[-1].copy()
pop_dens.loc[pop_dens.index[-1], numeric_cols] = new_row.values
pop_dens
```

```python
numeric_cols = pop_dens.columns[1:]
new_row = (pop_dens[numeric_cols].iloc[2] / pop_dens[numeric_cols].iloc[1])*100
pop_dens.loc[pop_dens.index[-1] + 1] = pop_dens.iloc[-1].copy()
pop_dens.loc[pop_dens.index[-1], numeric_cols] = new_row.values
pop_dens
```

```python
pop_dens.loc[55, 'Country'] = 'Difference'
pop_dens.loc[56, 'Country'] = 'Difference in %'
pop_dens
```

```python
pop_dens.round(0)
```

```python
pop_dens['Population'] = pop_dens['Population'].map('{:,.0f}'.format)
pop_dens['Area(km2)'] = pop_dens['Area(km2)'].map('{:,.0f}'.format)
```
