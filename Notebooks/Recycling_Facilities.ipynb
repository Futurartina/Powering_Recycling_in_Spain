```python
url = "http://www.geocountries.com/country/geolocation"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
response
```

```python
coordinates = soup.find_all("div", {"class": "table-responsive"})

list_ = [i.find_all("th") for i in coordinates]
columns = [i.getText() for j in list_ for i in j]
columns
```

```python
list_ = [i.find_all("td") for i in coordinates]
rows = [i.getText() for j in list_ for i in j]
rows_filtered = [i for i in rows if len(i) > 1]
rows_filtered
```

```python
countries = []
def create_custom_dict(sublist, keys):
    return {keys[i]: sublist[i] for i in range(len(sublist))}
# Defines a function to create a dictionary with custom keys
```

```python
key_names = ["Country", "Latitude", "Longitude", "DD", "DMS", "DDM"]
# Defines key names for each set of values.
```

```python
for i in range(0, len(rows_filtered), 6):
    sublist = rows_filtered[i:i+6]
    custom_dict = create_custom_dict(sublist, key_names)
    countries.append(custom_dict)   
# Loops to create dictionaries every 6 items.

print(countries)
# Prints the list of dictionaries.
```

```python
import pandas as pd
country = pd.DataFrame(countries)

country.drop(columns=["DD", "DMS", "DDM"], inplace=True)
country
# Creates a dataframe with the first 49 dictionaries.
```

```python
country.drop(country.index[-1], inplace=True)
country
```

```python
load_dotenv()
token = os.getenv("token")
```

```python
def requests_for_foursquare (query, lat, lon, radius):
    url = f"https://api.foursquare.com/v3/places/search?query={query}&ll={lat}%2C{lon}&radius={radius}"
    
    headers = {
        "accept": "application/json",
        "Authorization": token
        }  
    response = requests.get(url, headers=headers)

    return response.json()
```

```python
for index, row in country.iterrows():
    country_name = country["Country"]
    lat = row["Latitude"]
    lon = row["Longitude"]
```

```python
def venue_count(df, queries, radius):
    df_copy = df.copy()

    for query in queries:
        column_name = f'{query}_count'
        empty_list = []

        for index, row in df_copy.iterrows():
            lat = row["Latitude"]
            lon = row["Longitude"]

            try:
                result = requests_for_foursquare(query, lat, lon, radius)
                if "results" in result:
                    empty_list.append(len(result['results']))
                else:
                    print(f"Error in response for {lat}, {lon} with query: {query}")
            except requests.RequestException as e:
                print(f"Error: {e} for {lat}, {lon} with query: {query}")

        df_copy[column_name] = empty_list

    return df_copy
```

```python
queries = ['Recycling Facility', 'Plastics Supplier', 'Renewable Energy Service', 'Waste Management Service', 'Water Treatment Service', 'Fast Food Restaurant', 'Vegan and Vegetarian Restaurant', 'Steakhouse', 'Renewable Energy Service', 'Used Bookstore', 'Organic Grocery', 'Public Transportation']
dataframe = venue_count(country, queries, 30000)
```
